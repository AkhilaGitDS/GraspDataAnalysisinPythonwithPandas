Tabular data is just data, that looks like table meaning data with rows and columns, which also looks like excel spreadsheet.
Common formats like CSV(comma separated value) is also a tabular data. Can also easily read a tabular data file into pandas, even directly from a URL!

```python
import pandas as pd
```

```python
chipotle = pd.read_table('http://bit.ly/chiporders')
```

```python
The above table is read flawlessly from url http://bit.ly/chiporders, since the table was formatted perfectly.
```

```python
chipotle.head()
```

The above was bit easy case, lets take a stronger case:

```python
pd.read_table('http://bit.ly/movieusers')
```

From the above we can say that something is really wrong(it put everything in one column). This above is not looking like table.From the above output we can say like Pipe separators are separating the fields which is the problem. So here how can i say that about pipe separator to 'read_table'. So pandas.read_table has second argument which is 'sep' and its the delimiter to use and the default is 'tab' character. so here we need to rather tell to pandas our separator is 'pipe' character.

```python
pd.read_table('http://bit.ly/movieusers', sep='|')
```

Above one after adding second argument sep='|' looks lot better than before, as we can see now each of the fields are in their own column. we have one more problem is first row is converted as header row which is not a header row, which is just row of data. Here we need to say pandas there is no Header, as below:

```python
pd.read_table('http://bit.ly/movieusers', sep='|', header=None)
```

Above looks great, data is in data section of Dataframe. Actually we can define column names which is just integer numbers now.

Below now we will create a Python list called 'cols' which is list of strings and by seeing the data i will name the columns.

```python
cols = ['user_id','age','gender','profession','postal_code']
```

And will pass above 'cols' to read_table, as the value of argument 'names'. so the pandas know to use those strings as the columns names.

```python
pd.read_table('http://bit.ly/movieusers', sep='|', header=None, names=cols)
```

Now above table looks perfect after adding 'names', 'header', 'sep' as argument in read_table. its ready to work with and can save to a object like 'users' as dataframe below.

```python
users = pd.read_table('http://bit.ly/movieusers', sep='|', header=None, names=cols)
```

This is how you work with read_table, if you have tabular data try reading it. Look what happens, if it doesn't look right just iterate the process by adjusting the arguments in read_table. pls refer to read_table documentation which will help you to figure out which arguments you need to alter in order to look right.

###### Super Usefull Addon: 2 arguments in read_table, one is "skiprows" and "skipfooter". what do we use these arguments for?
Lets imagine you have a data file and someone has added some text or notes on top or bottom of the file, which is not real data but they are just some notes about file which they added in file itself. You can use these "skiprows" and "skipfooter" argument and tell pandas to skip certain rows, so that pandas can focus on real data in the file.
