Lets see how to display the rows of a DataFrame which have a certain column value. 

```python
import pandas as pd
```

```python
movie = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movie.head()
```

Above each row mean a movie with some more information about that.

```python
movie.shape
```

###### Will try to pull only movies which has longer duration as below:

Will create a python list of bool(type boolean)that has same exact length as Dataframe which says 'True' if duration is equal to or greater than 200 else 'False'.Python list says which row matches given condition. 

Will do this using regular python 'for' loop:

```python
type(True)
```

```python
type(False)
```

```python
booleans=[] #create a python list called 'booleans' which started empty
for term in movie.duration: #'movie.duration' is a pandas series which can be iterable like a python list
    if term >=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:10]
```

```python
len(booleans)
```

Above boolean length is exactly same as dataframe rows.

So till now we have created a list of boolean which has same length as dataframe and successfully says which rows match the condition.

Next, we will convert list of boolean to pandas series.

```python
long_run = pd.Series(booleans) # This line converts booleans into pandas Series.
```

```python
long_run.head() #All methods of Dataframe can be used with Series as well.
```

```python
movie[long_run]  # Here will pass'long_run' series to 'movie' dataframe using bracket notation.
```

In Above command, we passed Series(long_run) of 'True's and 'False's, look how pandas responding by showing us 'movie' Dataframe with all columns but only showing the rows in which 'duration' is at the minimum 200 minutes.

This is bit more steps to follow in pandas.

###### More simple ways to do above work in pandas:

```python
long_run
```

```python
long_run = movie.duration >= 200 # 1st Simple method and this line completely replaces for loop which is used before.
long_run.head()
```

" movie.duration >= 200 " : Here Series name with comparision operator, pandas know that we are not looking for single 'True' or 'False'. pandas know that we expecting comparision against each element in the series and returns a Series of 'True's and 'False's. 

```python
movie[long_run]
```

Next step instead of creating a separate Series 'long_run', we can write the filter condition which we used to create 'long_run' directly in the 'movie' Dataframe as below:

```python
movie[movie.duration >= 200] # 2nd Simple method
```

So , Above are the totally 3 ways where we can pick rows of pandas dataframe by column values based on some conditions. Very recent method or line of code is simple " movie[movie.duration >= 200] " and most useful where code within brackets create Series of boolean with same length as Dataframe and says which rows we want to select in Dataframe. 

###### Useful hint: Only pick genre of the long duration of the movies

And not to care about rest of the columns.

```python
movie[movie.duration >= 200] # This line of code is generating a Dataframe
```

```python
movie[movie.duration >= 200].genre # Pulling only 'genre' Series from " movie[movie.duration >= 200]" Dataframe.
```

### Afterall, the best practice is to use " loc method "

```python
movie.loc[movie.duration >= 200]
```

.loc[] is primarily label based, but may also be used with a boolean array.

```python
movie.loc[movie.duration >= 200, 'genre'] # loc is saying, need rows from duration atleast 200mins for 'genre' column
```

```python
movie.loc[movie.duration >= 200, ['genre','title']] # loc is saying, need rows from duration atleast 200mins for 'genre' and 'title' column
```

This is best practice how we can filter dataframe by condition and pulling out particular columns from that results.
