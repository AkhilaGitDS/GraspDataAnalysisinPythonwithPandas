lets dig in more about right way to pick rows and columns out of Dataframe. pandas gives us great number of choices for doing so. Best practices for row and column selection are using the loc, iloc methods are very flexible and powerful.

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head() # View first few(default:5) rows of data
```

###### Loc[ ]  method: 

Loc[] is a Dataframe method for filtering rows and selecting columns by label, by label means index for columns(column name). In brief, loc[] is selecting things by label.Will use bracket[] instead of parenthesis ().  

Loc [] command format is what rows do we need, what column do we need as below :

```python
ufo.loc[0,:] # 0 rows and : (means all) all columns
```

Here we got a pack of Series which is just showing us first row of the Dataframe (row 0) and all columns(which is all 5 columns).

How to choose multiple rows, which is shown below :    

```python
ufo.loc[[0,1,2],:] #Pass a list what all rows required and all column required
```

Here above we got 3 rows of data with all columns

Since we have continuos block of rows(0,1,2) we can write above command in most efficient way as below :

```python
ufo.loc[0:2, :] # 0 through 2 and all columns
```

To remember: Here [0:2] means inclusive of both 0 and 2 and all the mid numbers with whole rows. 

```python
ufo.loc[0:2]
```

This again gives same above 2 results because if nothing is passed in row or column location pandas assume by default as all rows or all columns. But above command is not recommended because Python philosophy is explicit is better than implicit.

###### We saw row selection using loc[ ], next will look into column selection using loc[ ]

```python
ufo.loc[:,['City']] # : - All rows and column 'City'.
```

```python
ufo.loc[:,['City','State']] # :(All Rows) and multiple Columns
```

```python
ufo.loc[:,'City':'State']
```

(:)-All Rows and City through State columns (inclusive of both City and State and all other middle columns)

We can offcourse combine these two in terms of Row selection and Column Selection as below:

```python
ufo.loc[0:2,'City':'State'] 
```

Rows 0 through 2 and Columns City through State, which is very flexible.

What is the other way we can do the above without using loc[] method?  :

```python
ufo.head(3).drop('Time',axis=1)
```

Above 2 results are same but 2 different commands.

Point to Remember: There is always lots of ways to do same thing in pandas and it comes down to what methods we feel most comfortable.

.loc[ ] is super powerful and flexible.

###### loc[ ] with boolean conditions: 

```python
ufo[ufo.City=='Oakland']
```

We can do the above same with .Loc[] with even bit more flexible.

```python
ufo.loc[ufo.City=='Oakland',:]
```

```python
ufo.loc[ufo.City=='Oakland', 'State':]
```

```python
ufo[ufo.City=='Oakland'].State
```

Above ufo.loc[ufo.City=='Oakland', 'State':]  and ufo[ufo.City=='Oakland'].State commands gives same results for the column 'State'but there is subtle difference on how code with loc[] and without loc[] executes, command without loc[] is known as 'Chain Indexing' and cause problems under certain scenario's. But command with loc[] won't cause problems. 

So if we are doing a boolean selection of rows then selection of columns, command with loc[] accomplishes it in one internal operation but command without loc[] takes two operations because of this reason command with loc[] is safer to execute the command the way we want.

###### iLoc[ ] method:

iloc[] is used for filtering rows and selecting columns by integer position, thats what the "i" stands for is integer position.

```python
ufo.iloc[:,[0,3]] # Require all(:) rows and columns in position 0 and 3
# command format looks same as loc[], like what rows and columns do we need
```

Here we will get City and State column because City column position is 0 and State column position is 3 in the Dataframe ufo.

```python
ufo.iloc[:,0:4] # We can also use Range, for instance here 0:4 
```

```python
ufo.columns
```

Here ufo.iloc[:,0:4] command, it shows you the columns in the position 0,1,2 & 3. So with iloc[], its exclusive of 2nd number(position 4 column) and inclusive of 1st number(position 0 column).

```python
list(range(0,4))
```

iloc[] is just like list(range()).

So when we using iloc[], its exclusive of 2nd number position value in command and in loc[], inclusive of both number position value which is really important to remember.

```python
ufo.iloc[0:3, :] # Here we want rows 0,1 and 2 and all columns
```

Here it did not include row 3.

So the big difference between loc[] and iloc[] is, loc[] for Labels and iloc[] for integer position. loc[] when we do ranges its inclusive on both sides and iloc[] inclusive of the start and exclusive of the stop. 

###### Some shortcuts we dont need to use them but we should be aware of them:

a> So one of them is if we have learned before how to select multiple columns as below: 

```python
ufo[['City','State']]
```

Here ufo[['City','State']] basically the inner brackets[] represent a list which is list of strings.
If we pass a list of strings to the outer brackets [], it knows that you want to select columns.
But it looks like lot of logic to remember here.

Instead of above, just use loc[] as below :

```python
ufo.loc[:,['City','State']] # loc[] with which rows and which columns
```

Above with loc[] is recommended because its very clear what we want.

Often seen people using below way, we might think that is referring to rows or columns or labels or integers which is confusing:

```python
ufo[0:2]
```

Here what we got when we ran it, its actually referring to rows and by integer position where the logic in command is not clear. 

Please always stay away from if logic is not clear enough and its confusing.

```python
ufo.iloc[0:2, :]
```

Command with iloc - ufo.iloc[0:2, :] is much better because we know rows position 0 till 2 is inclusive but 2 is exclusive and all columns.
